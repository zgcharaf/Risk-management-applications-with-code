## VAR 10% parametric approach using SMA - Normal times

```python

# Calculate SMA of daily volatility
sma_daily_volatility = portfolio_daily_returns_reference.rolling(window=20).std()

# Z-score for 10% VaR under normal distribution assumption
z_score_10_percent = -1.28  # For the 10th percentile

# Calculate 1-day 10% VaR using the SMA of volatility
VaR_10_1_day_SMA = z_score_10_percent * sma_daily_volatility * portfolio_value

# Taking the last value to get the most recent estimate
VaR_10_1_day_value = VaR_10_1_day_SMA.dropna().iloc[-1]

print(f"1-day 10% VaR using SMA is: {VaR_10_1_day_value:.2f} EUR")

# Assuming you want to plot the daily returns and highlight the VaR
# Ensure portfolio_daily_returns_reference is defined before this step
plt.figure(figsize=(10, 6))
sns.histplot(portfolio_daily_returns_reference, bins=50, kde=True, color='blue', label='Daily Returns')
plt.axvline(VaR_10_1_day_value / portfolio_value, color='red', linestyle='--', label=f'10% VaR: €{VaR_10_1_day_value:.2f}')
plt.title('Histogram of Daily Returns with 10% VaR')
plt.xlabel('Daily Return')
plt.ylabel('Frequency')
plt.legend()
plt.show()
```

## Parametric approach using EWMA normal times 

```python
# Parameters and initial settings
lambda_ = 0.5

def calculate_ewma_volatility(daily_returns, lambda_):
    ewma_variance = np.zeros_like(daily_returns)
    ewma_variance[0] = np.var(daily_returns)  # Initial variance
    
    for t in range(1, len(daily_returns)):
        ewma_variance[t] = lambda_ * ewma_variance[t-1] + (1 - lambda_) * daily_returns[t-1]**2
    
    ewma_volatility = np.sqrt(ewma_variance)
    return ewma_volatility

# Calculate EWMA volatility for both portfolios
ewma_volatility_normal = calculate_ewma_volatility(portfolio_daily_returns_reference, lambda_)

# Calculate the last day EWMA volatility as the representative volatility
ewma_volatility_normal_last = ewma_volatility_normal[-1]

# Calculate daily 10% VaR for both portfolios
daily_var_10_normal = z_score * ewma_volatility_normal_last * portfolio_value

plt.figure(figsize=(10, 6))
sns.histplot(portfolio_daily_returns_reference, bins=30, kde=True, color="blue", label="Daily Returns")
plt.axvline(daily_var_10_normal / portfolio_value, color='red', linestyle='--', label=f'10% VaR: €{daily_var_10_normal:.2f}')
plt.title("Histogram of Historical Daily Returns with 10% VaR")
plt.xlabel("Daily Return")
plt.ylabel("Frequency")
plt.legend()
plt.show()

# Print the 10% VaR result
print(f"The 10% VaR calculated using the non-parametric approach is: {z_score * ewma_volatility_normal_last *100:.2f}%")
print(f"In monetary terms, for a portfolio value of €10,000, the 10% VaR is approximately: €{-z_score * ewma_volatility_normal_last * 10000:.2f}")


```

## Monte Carlo Parametric approach 

###

```python
 
# Number of simulations
n_simulations = 100000
portfolio_value_reference=10000
portfolio_value_stress=10000
# Parametric Monte Carlo Simulation
import matplotlib.pyplot as plt

# Function to perform parametric VaR calculation and plotting
def parametric_VaR_with_plot(returns, portfolio_value, confidence_level=10):
    mean_returns = returns.mean()
    std_returns = returns.std()
    
    # Simulate future returns
    simulated_returns = np.random.normal(mean_returns, std_returns, n_simulations)
    
    # Calculate simulated end portfolio values
    simulated_end_values = portfolio_value * (1 + simulated_returns)
    
    # Calculate VaR
    VaR = np.percentile(simulated_end_values - portfolio_value, confidence_level)
    
    # Plotting
    plt.figure(figsize=(10, 6))
    plt.hist(simulated_end_values - portfolio_value, bins=50, alpha=0.75)
    plt.axvline(x=VaR, color='r', linestyle='--', label=f'10% VaR: €{VaR:.2f}')
    plt.title('Simulated Portfolio End Values Distribution (Parametric)')
    plt.xlabel('Portfolio Value Change')
    plt.ylabel('Frequency')
    plt.legend()
    plt.show()
    
    return VaR

# Parametric VaR Calculation and Plotting for both portfolios
VaR_10_percent_reference_parametric = parametric_VaR_with_plot(portfolio_daily_returns_reference, portfolio_value_reference)
VaR_10_percent_stress_parametric = parametric_VaR_with_plot(portfolio_daily_returns_stress, portfolio_value_stress)

print(f"Reference Portfolio 10% VaR (daily, parametric): €{VaR_10_percent_reference_parametric:.2f}")
print(f"Stress Portfolio 10% VaR (daily, parametric): €{VaR_10_percent_stress_parametric:.2f}")
```

## Monte Carlo Non-Parametric approach 

```python
# Non-parametric Monte Carlo Simulation
# Function to perform non-parametric VaR calculation and plotting
def non_parametric_VaR_with_plot(returns, portfolio_value, confidence_level=10):
    # Sample from historical returns
    simulated_returns = np.random.choice(returns, size=n_simulations, replace=True)
    
    # Calculate simulated end portfolio values
    simulated_end_values = portfolio_value * (1 + simulated_returns)
    
    # Calculate VaR
    VaR = np.percentile(simulated_end_values - portfolio_value, confidence_level)
    
    # Plotting
    plt.figure(figsize=(10, 6))
    plt.hist(simulated_end_values - portfolio_value, bins=50, alpha=0.75)
    plt.axvline(x=VaR, color='r', linestyle='--', label=f'10% VaR: €{VaR:.2f}')
    plt.title('Simulated Portfolio End Values Distribution (Non-parametric)')
    plt.xlabel('Portfolio Value Change')
    plt.ylabel('Frequency')
    plt.legend()
    plt.show()
    
    return VaR

# Non-parametric VaR Calculation and Plotting for both portfolios
VaR_10_percent_reference_non_parametric = non_parametric_VaR_with_plot(portfolio_daily_returns_reference, portfolio_value_reference)
VaR_10_percent_stress_non_parametric = non_parametric_VaR_with_plot(portfolio_daily_returns_stress, portfolio_value_stress)

print(f"Reference Portfolio 10% VaR (daily, non-parametric): €{VaR_10_percent_reference_non_parametric:.2f}")
print(f"Stress Portfolio 10% VaR (daily, non-parametric): €{VaR_10_percent_stress_non_parametric:.2f}")
```

## Expected Shortfall

```python
import numpy as np
from scipy.stats import norm

portfolio_value = 10000  # Example value, adjust as necessary

# Function to calculate non-parametric and parametric ES
def calculate_ES(returns, portfolio_value):
    # Non-parametric ES
    sorted_returns = np.sort(returns)
    cutoff_index = int(np.ceil(5/100 * len(sorted_returns)))  # 5th percentile index
    ES_5_non_parametric = sorted_returns[:cutoff_index].mean()  # Mean of the lowest 5%
    
    # Parametric ES
    mean_return = returns.mean()
    std_dev = returns.std()
    z_score_5_percent = norm.ppf(0.05)  # Z-score for 5th percentile
    ES_5_parametric = mean_return + std_dev * z_score_5_percent
    ES_5_parametric_monetary = ES_5_parametric * portfolio_value
    
    return ES_5_non_parametric, ES_5_parametric, ES_5_parametric_monetary

# Calculate ES for the reference portfolio
ES_5_non_parametric_ref, ES_5_parametric_ref, ES_5_parametric_monetary_ref = calculate_ES(portfolio_daily_returns_reference, portfolio_value)
print(f"Reference Portfolio - Non-parametric Expected Shortfall (5th percentile): {ES_5_non_parametric_ref*100:.2f}%")
print(f"Reference Portfolio - Parametric Expected Shortfall (5th percentile): {ES_5_parametric_ref*100:.2f}%")
print(f"In monetary terms, for the reference portfolio, the 5% ES is approximately: €{ES_5_parametric_monetary_ref:.2f}")

# Calculate ES for the high stress portfolio
ES_5_non_parametric_stress, ES_5_parametric_stress, ES_5_parametric_monetary_stress = calculate_ES(portfolio_daily_returns_stress, portfolio_value)
print(f"High Stress Portfolio - Non-parametric Expected Shortfall (5th percentile): {ES_5_non_parametric_stress*100:.2f}%")
print(f"High Stress Portfolio - Parametric Expected Shortfall (5th percentile): {ES_5_parametric_stress*100:.2f}%")
print(f"In monetary terms, for the high stress portfolio, the 5% ES is approximately: €{ES_5_parametric_monetary_stress:.2f}")
```

The difference between non-parametric and parametric ES in both portfolios highlights the importance of considering tail risks. The non-parametric method, which does not assume a specific distribution of returns, captures the risk of extreme loss more effectively, particularly in stressed market conditions.
